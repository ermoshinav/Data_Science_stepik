```python
import plotly.figure_factory as ff # нужен ипорт для отрисовки 4 гистограмм на одной картинке
import pandas as pd
```

```python
# подгрузили датафрейм
iris=pd.read_csv('https://stepik.org/media/attachments/course/4852/iris.csv',index_col=0)
group_labels = []    # Тут будет лежать названия колонок для легенды
hist_data = []       # Тут будут лежать сами эти колонки
```

```python
# В цикле наполним переменные выше значениями
for column in iris:
    group_labels.append(column)
    hist_data.append(iris[column])
```

```python
fig = ff.create_distplot(
    hist_data,
    group_labels
)

fig.show()
```

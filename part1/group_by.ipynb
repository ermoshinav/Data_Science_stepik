1. Пересчитаем число ног у героев игры **Dota2!** Сгруппируйте героев из датасэта по числу их ног (колонка ***legs***), и заполните их число в задании ниже.

```python
import pandas as pd
dota2 = pd.read_csv('https://stepik.org/media/attachments/course/4852/dota_hero_stats.csv')
```

```python
dota2.legs.value_counts()
```

Сгруппируйте по колонкам *attack_type* и *primary_attr* и выберите самый распространённый набор характеристик.

```python
dota2.groupby(['attack_type', 'primary_attr'], as_index=False).aggregate({'id':'count'}).sort_values('id', ascending=False)
```

Чему равны минимальная, средняя и максимальная концентрации аланина *(alanin)* среди видов рода *Fucus*. Округлите до 2-ого знака, десятичным разделителем является точка.

```python
import numpy as np
bio= pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
```

```python
np.around(bio.groupby('genus').alanin.describe().loc['Fucus', ['min',  'mean',  'max']].values, decimals=2)
```
